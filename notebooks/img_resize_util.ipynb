```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
img_size=256
img_extn='.jpg'
base_path='../data/Flower_Data/test'
imgs={'iris':4,'poppy':3,'rose':6,'waterlilly':2}
save_path=base_path+'/resized'
orig_path=base_path+'/original'
```

```python
def create_paths(base_path,img_extn,imgs):
    ret=[]
    for subj,num in imgs.items():
        for i in range(1,num+1):
            path=base_path+'/'+subj+str(i)+img_extn
            ret.append(path)
    return ret
```

```python
paths=create_paths(orig_path,img_extn,imgs)
paths
```

```python
img_bgr=cv2.imread(paths[0])
img_rgb=cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) 
resized_img=cv2.resize(img_rgb,(img_size,img_size))
plt.imshow(resized_img)
print(paths[0].split('/')[-1])
```

```python
def resize_imgs(paths,save_path):
    for path in paths:
        img_bgr=cv2.imread(path)
        #img_rgb=cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) 
        resized_img=cv2.resize(img_bgr,(img_size,img_size))
        cv2.imwrite(save_path+'/'+path.split('/')[-1], resized_img)
```

```python
resize_imgs(paths,save=True,save_path)
```
